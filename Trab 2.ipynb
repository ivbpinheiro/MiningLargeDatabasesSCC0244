```python
%%capture
%%sql

DROP TABLE IF EXISTS Tudo CASCADE;
CREATE TABLE Tudo(
    ID_PACIENTE        VARCHAR(40),
    CD_MUNICIPIO VARCHAR(100),
    DE_EXAME       VARCHAR(8000),
    HOSPITAL TEXT
    );
```

```python
%%capture
%%sql

INSERT INTO Tudo 
(ID_PACIENTE,DE_EXAME,HOSPITAL)
SELECT ID_PACIENTE,DE_EXAME,HOSPITAL
FROM hc_Exames;

UPDATE Tudo
SET CD_MUNICIPIO=pa.CD_MUNICIPIO
FROM hc_Paciente pa
WHERE Tudo.ID_PACIENTE=pa.ID_PACIENTE AND Tudo.HOSPITAL=pa.HOSPITAL;

INSERT INTO Tudo
(ID_PACIENTE,DE_EXAME,HOSPITAL)
SELECT ID_PACIENTE,DE_EXAME,HOSPITAL
FROM bpsp_Exames;

UPDATE Tudo
SET CD_MUNICIPIO=pa.CD_MUNICIPIO
FROM bpsp_Paciente pa
WHERE Tudo.ID_PACIENTE=pa.ID_PACIENTE AND Tudo.HOSPITAL=pa.HOSPITAL;

INSERT INTO Tudo
(ID_PACIENTE,DE_EXAME,HOSPITAL)
SELECT ID_PACIENTE,DE_EXAME,HOSPITAL
FROM fleury_Exames;

UPDATE Tudo
SET CD_MUNICIPIO=pa.CD_MUNICIPIO
FROM fleury_Paciente pa
WHERE Tudo.ID_PACIENTE=pa.ID_PACIENTE AND Tudo.HOSPITAL=pa.HOSPITAL;

INSERT INTO Tudo
(ID_PACIENTE,DE_EXAME,HOSPITAL)
SELECT ID_PACIENTE,DE_EXAME,HOSPITAL
FROM hsl_Exames;

UPDATE Tudo
SET CD_MUNICIPIO=pa.CD_MUNICIPIO
FROM hsl_Paciente pa
WHERE Tudo.ID_PACIENTE=pa.ID_PACIENTE AND Tudo.HOSPITAL=pa.HOSPITAL;

UPDATE TUDO SET cd_municipio='' WHERE cd_municipio='MMMM';
```

```python
%%sql

SELECT DE_EXAME as Exame, HOSPITAL, CD_MUNICIPIO as cidade, COUNT(*) Total
FROM TUDO
WHERE (DE_EXAME LIKE '%Corona%'
       or DE_EXAME LIKE '%Cov%'
      or DE_EXAME LIKE'%COV%'
      or DE_EXAME LIKE '%CORONA%'
      and DE_EXAME NOT LIKE '%Cor43%'
      and DE_EXAME NOT LIKE '%HKU%'
      and DE_EXAME NOT LIKE '%Cor229%'
      and DE_EXAME NOT LIKE '%Cor63%')
GROUP BY GROUPING SETS ((DE_EXAME,CD_MUNICIPIO, HOSPITAL),(HOSPITAL)) 
    ORDER BY HOSPITAL ASC,Exame, CD_MUNICIPIO ASC;
```
