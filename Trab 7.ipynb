# VGG16 - Extração de características de imagens

```python
import os
import pandas as pd
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np

# Carregando o modelo VGG16 pré-treinado
model = VGG16(weights='imagenet', include_top=False)  # Carrega o modelo sem a camada densa no topo

# Diretório onde estão localizadas as imagens
diretorio_imagens = '/content/pasta_de_destino/BoWFireDataset/train'  # Substitua pelo caminho do seu diretório

# Lista para armazenar os resultados das características das imagens
resultados = []

# Lista todos os arquivos no diretório
imagens = os.listdir(diretorio_imagens)

# Loop através de cada imagem no diretório
for img_nome in imagens:
    if img_nome.endswith('.jpg') or img_nome.endswith('.jpeg') or img_nome.endswith('.png'):
        # Caminho completo da imagem
        caminho_img = os.path.join(diretorio_imagens, img_nome)

        # Carregando e redimensionando a imagem
        img = image.load_img(caminho_img, target_size=(224, 224))  # Redimensiona a imagem para o tamanho esperado pela VGG16
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)

        # Obtendo as características da imagem usando a VGG16
        features = model.predict(x)

        # Armazenando as características em um dicionário
        resultado_imagem = {
            'NAME_IMG': img_nome,
            'CHARACTERISTICS_RAW': features.flatten()
        }

        resultados.append(resultado_imagem)

df_raw = pd.DataFrame(resultados)

print(df_raw)
```
